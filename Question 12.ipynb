## Highly divisible triangular number
*[Question 12]*


The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be $1 + 2 + 3 + 4 + 5 + 6 + 7 = 28$. The first ten terms would be:

> $1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...$

Let us list the factors of the first seven triangle numbers:

> **1:** $1$
<br>
**3:** $1,3$
<br>
**6:** $1,2,3,6$
<br>
**10:** $1,2,5,10$
<br>
**15:** $1,3,5,15$
<br>
**21:** $1,3,7,21$
<br>
**28:** $1,2,4,7,14,28$

We can see that $28$ is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?
____

**Working out:**
<br>
<img src="https://i.imgur.com/dqOmjYp.png" width="600" height="400">

```python
import math

def num_factors(num):
    # Get all the possible pairs of factors i.e. 12: [(1, 12), (2, 6), (3, 4)]
    factor_pairs = [(factor, int(num / factor)) for factor in range(1, int(math.sqrt(num)) + 1) if num % factor == 0]
    ans = len(factor_pairs * 2)
    # Do not count (sqrt, sqrt) as two unique factors
    if int(math.sqrt(num)) ** 2 == num: ans -= 1
    return ans
```

```python
i = 0
while True:
    i += 1
    # triangle number = 1 + 2 + 3 + ... + i = i / 2 * (i + 1)
    triangle_num = i / 2 * (i + 1)
    if num_factors(triangle_num) > 500:
        print(triangle_num)
        break
```
